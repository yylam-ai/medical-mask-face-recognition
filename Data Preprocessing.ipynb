# Process Data

```python
import cv2
from PIL import Image
import glob
import numpy as np
import pandas as pd
```

```python
tmp_full_data_sorted = []
y = []

for i in range(1, 162):
  for j in range (1, 501):
    tmp_full_data_sorted.append("%s" %i + "/" + "%s" %j + ".jpg")
    y.append(i)
```

```python
X = []

for image in tmp_full_data_sorted:
    with open(image, 'rb') as file:
        reshaped_img = np.array([])
        img = (cv2.imread(image, 0))
        reshaped_img = np.reshape(img, img.shape[0]*img.shape[1])
        transposed_img = np.array([reshaped_img]).T
        X.append(transposed_img)
```
